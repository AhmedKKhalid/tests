```python
import keyboard
from selenium import webdriver
import pandas as pd
import warnings 
import time
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(executable_path=r"C:\Users\int3\Desktop\drivers\chromedriver - Copy (2).exe")
```

```python
url= f'https://www.tripadvisor.com/Hotel_Review-g294201-d300669-Reviews-Ramses_Hilton-Cairo_Cairo_Governorate.html'
driver.get(url)
```

```python
df=pd.read_excel(r"C:\Users\int3\Desktop\All Data\scraping.xlsx",sheet_name="Sheet3")
df.head()
```

```python
df=df[36725:].reset_index()
```

```python
df.head()
```

```python
hName=[]
hReviews=[]
hRank=[]
hCategory=[]
hJp=[]
```

```python
for i in range(df.shape[0]):
    hotel_name="" ;reviews="";rank="";category=""
    try:
        if keyboard.is_pressed('q'):
            break 
        search_input=driver.find_element_by_class_name('qjfqs._G.B-.z._J.Cj.R0')
        search_input.clear()
        search_input.send_keys(df['Hotel Name'][i])
        time.sleep(7)
        search_button=driver.find_element_by_class_name('GzJDZ.w.z._S._F.Wc.Wh.Q.B-._G').click()
        time.sleep(10)
        try:
            hotel_name=driver.find_element_by_tag_name('h1').text
        except:
            hotel_name="NA"
        try:    
            try:
                reviews=driver.find_element_by_class_name('qqniT').text
            except:
                reviews=driver.find_element_by_class_name('AfQtZ').text
        except:
            reviews ="NA"   
        try:   
            try:
                rank=driver.find_element_by_class_name('cGAqf').text
            except:
                rank=driver.find_element_by_class_name('AYHFM').text                
        except:
            rank="NA"
        try:    
            try:
                category=driver.find_element_by_class_name('uwJeR').text
            except:
                category=driver.find_element_by_class_name('ZDEqb').text           
        except:
            category="NA"
    except:
        time.sleep(300)
        driver.get(url)
        
    hName.append(hotel_name)
    hReviews.append(reviews)
    hRank.append(rank)
    hCategory.append(category)
    hJp.append(df['JP Code'][i])
    print(df['JP Code'][i],";",hotel_name,";",reviews,",",rank,";",category)
```
